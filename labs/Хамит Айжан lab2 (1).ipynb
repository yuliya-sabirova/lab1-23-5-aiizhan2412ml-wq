```python
#1.1 Создайте одномерный массив из чисел от 1 до 20 включительно. Выведите его форму
import numpy as np
arr = np.arange(1, 21)
print("Массив:", arr)
print("Форма массива:", arr.shape)
# Форма массива (shape) показывает размеры массива по всем измерениям, а его длина (len) — количество элементов только вдоль первой оси.
```

```python
#1.2 Преобразуйте массив в матрицу размером 5x4. Проверьте результат.
import numpy as np
arr = np.arange(1, 21)
matrix = arr.reshape(5, 4)
print("Матрица 5x4:")
print(matrix)
print("Форма матрицы:", matrix.shape)
# Если число элементов в массиве не соответствует новой форме, функция reshape() вызовет ошибку
```

```python
#1.3 Увеличьте размерность массива до 20x1 с помощью np.newaxis и выведите форму нового массива.
import numpy as np
arr = np.arange(1, 21)
arr_reshaped = arr[:, np.newaxis]
print("Массив размером 20x1:")
print(arr_reshaped)
print("Форма массива:", arr_reshaped.shape)
# Увеличение размерности массива нужно для обеспечения совместимости с другими массивами 
#при математических операциях и подготовке данных к нужному формату, например, в машинном обучении.
```

```python
#2.1 Создайте массив из 12 равномерно распределенных значений в диапазоне [0, 5] с помощью np.linspace(). Преобразуйте его в матрицу 3x4.
import numpy as np
arr = np.linspace(0, 5, 12)
matrix = arr.reshape(3, 4)
print("Матрица 3x4 из равномерных значений:")
print(matrix)
# Линейное распределение создаёт равномерно расположенные точки на заданном интервале, 
#а случайное распределение генерирует точки в этом интервале в произвольном порядке без равномерности.
```

```python
#2.2 Сгенерируйте матрицу 3x3 с: Случайными числами из равномерного распределения [0, 1).
# Случайными числами из нормального распределения. Целыми числами в диапазоне [1, 10].
import numpy as np

# Случайные числа из равномерного распределения [0, 1)
uniform_matrix = np.random.rand(3, 3)

# Случайные числа из нормального распределения (среднее=0, стандартное отклонение=1)
normal_matrix = np.random.randn(3, 3)

# Целые случайные числа в диапазоне [1, 10]
integer_matrix = np.random.randint(1, 11, size=(3, 3))

print("Матрица с равномерно распределёнными числами [0,1):\n", uniform_matrix)
print("\nМатрица с нормально распределёнными числами:\n", normal_matrix)
print("\nМатрица с целыми числами в диапазоне [1, 10]:\n", integer_matrix)
```

```python
# 2.3 Сформируйте массив из 6 чисел по логарифмической шкале в диапазоне [10^1, 10^3] и преобразуйте его в двумерный массив 2x3
import numpy as np
log_array = np.logspace(1, 3, 6)
log_matrix = log_array.reshape(2, 3)
print("Массив по логарифмической шкале 2x3:")
print(log_matrix)
# Линейное распределение равномерно распределяет значения по интервалу, 
#логарифмическое — равномерно по логарифмической шкале, а геометрическое — описывает вероятность числа испытаний до первого успеха в серии событий.
```

```python
# 3.1 Создайте диагональную матрицу 4x4 с элементами [1, 2, 3, 4] на главной диагонали.
import numpy as np
diag_elements = [1, 2, 3, 4]
diag_matrix = np.diag(diag_elements)
print("Диагональная матрица 4x4:")
print(diag_matrix)
# Диагональная матрица используется для удобного представления и вычисления масштабирования, 
#преобразований и для упрощения операций с матрицами, поскольку умножение на диагональную матрицу сводится к поэлементному умножению по диагонали.
```

```python
# 3.2 Создайте единичную матрицу размера 5x5.
import numpy as np
identity_matrix = np.eye(5)
print("Единичная матрица 5x5:")
print(identity_matrix)
# не изменяет умножаемую матрицу
```

```python
# 3.3 Постройте
import numpy as np

# Нижнетреугольная матрица из чисел 1-16 в форме 4x4
arr1 = np.arange(1, 17).reshape(4, 4)
lower_triangular = np.tril(arr1)

# Верхнетреугольная матрица из чисел 1-25 в форме 5x5
arr2 = np.arange(1, 26).reshape(5, 5)
upper_triangular = np.triu(arr2)

print("Нижнетреугольная матрица 4x4:")
print(lower_triangular)

print("\nВерхнетреугольная матрица 5x5:")
print(upper_triangular)
# np.tril() оставляет нижнюю треугольную часть матрицы, а np.triu() — верхнюю, обнуляя остальные элементы.
```

```python
# 4.1 Элементарные операции с матрицами
import numpy as np
m = int(input("Введите размер матриц m: "))
A = np.random.randint(1, 11, size=(m, m))
B = np.random.randint(1, 11, size=(m, m))
print("Матрица A:\n", A)
print("Матрица B:\n", B)
sum_matrix = A + B
print("\nСложение A + B:\n", sum_matrix)
sub_matrix = A - B
print("\nВычитание A - B:\n", sub_matrix)
mul_matrix = A * B
print("\nЭлементное умножение A * B:\n", mul_matrix)
div_matrix = A / B
print("\nЭлементное деление A / B:\n", div_matrix)
```

```python
# 4.2
sqrt_matrix = np.sqrt(A)
print("\nКвадратный корень каждого элемента A:\n", sqrt_matrix)
```

```python
# 4.3
transposed_B = B.T
print("\nТранспонированная матрица B:\n", transposed_B)
```

```python
# 5.1 Создайте матрицу 4x4, где каждый элемент вычисляется как сумма его индексов. Используйте np.fromfunction().
import numpy as np
# Создание матрицы 4x4, где каждый элемент равен сумме его индексов
matrix = np.fromfunction(lambda i, j: i + j, (4, 4), dtype=int)
print("Матрица 4x4, где каждый элемент — сумма его индексов:\n", matrix)
```

```python
# 5.2 Постройте матрицу 3x3, где каждый элемент равен произведению индексов строки и столбца.
import numpy as np
matrix = np.fromfunction(lambda i, j: i * j, (3, 3), dtype=int)
print("Матрица 3x3 (произведение индексов строки и столбца):\n", matrix)
```

```python
# 6.1 Анализ элементов матрицы
import numpy as np
matrix = np.random.rand(5, 5)
print("Матрица 5x5:\n", matrix)
total_sum = np.sum(matrix)
print("\nСумма всех элементов:", total_sum)
row_sums = np.sum(matrix, axis=1)
print("\nСумма по строкам:", row_sums)
column_sums = np.sum(matrix, axis=0)
print("\nСумма по столбцам:", column_sums)
```

```python
# 6.2 Анализ элементов матрицы
import numpy as np
matrix = np.random.rand(5, 5)
print("Матрица 5x5:\n", matrix)

# Сумма всех элементов
total_sum = np.sum(matrix)
print("\nСумма всех элементов:", total_sum)

# Сумма по строкам (axis=1)
row_sums = np.sum(matrix, axis=1)
print("\nСумма по строкам:", row_sums)

# Сумма по столбцам (axis=0)
column_sums = np.sum(matrix, axis=0)
print("\nСумма по столбцам:", column_sums)
```
